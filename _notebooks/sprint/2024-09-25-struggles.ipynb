# Struggles and Accomplishments in Building My Website

### Struggles ü•¥

#### 1. Hamburger Menu Placement Gone Wrong (Sept 12, 2024)
I thought I'd be clever by adding a hover-over hamburger menu for my PLL section. Turns out, getting it to look just right with HTML and CSS was a struggle. The icon wouldn't align properly, and when it did, the dropdown links scattered across the page like they had their own agenda. 

**Solution:** After a lot of trial and error, I tweaked the CSS positioning to `absolute` and ensured the links were inside the correct container element. Debugging CSS is like trying to herd cats!

#### 2. Markdown vs. HTML Chaos (Sept 13, 2024)
Markdown is supposed to be simple, right? Well, not when you're mixing it with HTML! I had a section where I needed more advanced formatting, so I threw in some HTML code. But then, Markdown decided it didn't want to play nice with the HTML elements. Paragraphs wouldn't align, and some random HTML tags appeared in the output. ü§¶‚Äç‚ôÇÔ∏è

**Solution:** I figured out that some Markdown parsers just don't like mixing HTML and Markdown, so I switched entirely to HTML for that section. Consistency wins!

#### 3. Styling Headaches with the `config.yml` File (Sept 14, 2024)
My page looked fine until I decided to mess with the `config.yml` file to adjust the theme. Suddenly, some fonts became unreadable, and the colors were not what I wanted. It was like my page was protesting the changes I was making. 

**Solution:** I dug deeper into how GitHub Pages handles theme skins and edited the `config.yml` to include only minor, safe changes. Lesson learned: small changes can make a big impact, for better or worse.

#### 4. JavaScript is Not Markdown-Friendly (Sept 15, 2024)
Trying to embed my Cookie Clicker game into a Markdown page was a headache. The Markdown page just wasn‚Äôt having it. It either showed the code itself or completely ignored it.

**Solution:** I wrapped the JavaScript in `<script>` tags directly in the HTML and added the game to a separate page linked from my Markdown. It's not ideal, but at least it works!

#### 5. Jekyll Liquid Syntax Errors (Sept 16, 2024)
While trying to implement some Liquid tags in my layout, I ran into errors that broke the whole page. It was frustrating because Liquid looks easy until you realize one typo can crash your site.

**Solution:** Careful debugging and reviewing Jekyll's documentation helped me fix the issues. Pro tip: double-check every curly brace and percentage symbol!
